```python
import yfinance as yf
import pandas as pd
import pickle
import numpy as np


ticker = "META"
stock = yf.Ticker(ticker)
dividends = stock.dividends  # Fetches historical dividend data
# dividends.tail(15)
dividends
```

```python
with open('new_list.pkl','rb') as file:
    load = pickle.load(file)
load
```

```python
load.append('META')
```

```python
# tickers = ['AA', 'AAL', 'AAPL', 'ABBV', 'ABNB', 'ABT', 'ACI', 'ACN','MSFT']
tickers = load

# Dictionary to store dividend data for each ticker
dividend_data = {}

# Fetching dividend data for each ticker
for ticker in tickers:
    stock = yf.Ticker(ticker)
    try:
        dividends = stock.dividends  # Fetch historical dividend data
        dividend_data[ticker] = dividends  # Store in dictionary
        print(f"Fetched dividends for {ticker}")
    except Exception as e:
        print(f"Failed to fetch data for {ticker}: {e}")

# Save the data to a pickle file
with open('dividend_dataAAAAA.pkl', 'wb') as file:
    pickle.dump(dividend_data, file)

print("All dividend data saved to 'dividend_data.pkl'.")
```
